```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/house_price.csv")
```

# Univariate Analysis
Analysis to one and only one feature only. So it includes:
1. Central Tendency: identify the center of the data distribution, finding the value that represent a group as a whole (mean, median, mode)
2. Dispersion: identify how spread the data in the distribution (variance, range, standard deviation, maximum, minimum)
3. Count and Null Count

```python
df.describe()
```

```python
# visualization
def plot_distribution(df, feature):
    sns.histplot(df[feature], kde=True, bins=int(np.ceil(np.log2(len(df))) + 1))

    mean_value = df[feature].mean()
    median_value = df[feature].median()
    mode_value = df[feature].mode()[0]
    plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.axvline(median_value, color='blue', linestyle='dashed', linewidth=2, label='Median')
    plt.axvline(mode_value, color='green', linestyle='dashed', linewidth=2, label='Mode')

    # Add legend
    plt.legend()

    # Add title and labels
    plt.title(f"Data Distribution {feature} with Mean, Median, and Mode")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    # Show the plot
    plt.show()

for c in df.select_dtypes(include=["int64", "float64"]).columns:
    plot_distribution(df, c)
```
