## 3. Univariate Analysis

### 3.1 Distribution 
Shows how often the data occurs in the dataset. We would also measure the shape of the distribution using skewness and kurtosis. Skewness is the measure of distortion that deviates from the symmetrical bell curve. Kurtosis is a measure tailedness of the distribution whether they have heavy tails or lighter tails. Heavy tails meaning that the distribution has more outliers than usual (normal distribution), while lighter tails has lower outliers than usual (normal distribution)

The value of skewness will mean different thing:
1. skewness = 0, indicates normal distribution
2. skewness > 0, indicates a right skewed distribution
3. skewness < 0, indicates a left skewed distribution

The value of kurtosis will mean different thing:
1. kurtosis = 0, indicates normal distribution (mesokurtic)
2. kurtosis > 0, indicates a right skewed distribution (leptokurtic)
3. kurtosis < 0, indicates a left skewed distribution (platykurtic)

#### Numerical

```python
# measure central tendency
skewness, kurtosis = [], []
for f in numerical_cols:
    skewness.append(df[f].skew()) # finding mean
    kurtosis.append(df[f].kurt()) # finding medain

pd.DataFrame({
    'feature': numerical_cols,
    'skewness': skewness,
    'kurtosis': kurtosis,
})
```

```python
def plot_distribution(data: pd.DataFrame, feature: str, title: str) -> None:
    # histogram plot
    fig, ax = plt.subplots(ncols=2, sharey=True, figsize=(10, 7))
    sns.histplot(data[feature], ax=ax[0])
    ax[0].set_title(title + f' (skewness = {data[feature].skew():.3f}, kurtosis = {data[feature].kurt()})')

    log_transformed_data = data[feature].apply(lambda x: np.log(x)) 
    sns.histplot(log_transformed_data, ax=ax[1])
    ax[1].set_title(title + f' (skewness = {log_transformed_data.skew():.3f}, kurtosis = {log_transformed_data.kurt()})')
    plt.show()


for f in numerical_cols:    
    # plot distribution
    plot_distribution(data=df.copy(), feature=f, title=f"Distribution of {f}")
    break
```

Some data has some extreme skewness into it we can use log transformation to reduce it

```python
box_cox_df = df[numerical_cols].apply(lambda x: np.log(x), axis=0)
for f in numerical_cols:    
    # plot distribution
    plot_distribution(data=box_cox_df, feature=f, title=f"Distribution of {f}")
```

The skewness of the data is reduced

### 3.2 Central Tendency
This part is useful for describing the "tendency" of data. It means what is the value where most of the data tend to cluster.

#### Numerical

```python
def plot_tendency(data: pd.DataFrame, feature: str, title: str) -> None:
    # calculate mean, median, mode
    mean = data[feature].mean()
    median = data[feature].median()
    mode = data[feature].mode().values[0]
    fig, ax = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (0.2, 1)})
    title += f'(mean={mean:.3f}, median={median:.3f}, mode={mode:.3f})'
    fig.suptitle(title)
    # box plot
    sns.boxplot(data=data, x=feature, ax=ax[0])
    ax[0].axvline(mean, color='r', linestyle='--')
    ax[0].axvline(median, color='g', linestyle='-')
    ax[0].axvline(mode, color='b', linestyle='-.')
    # histogram plot
    sns.histplot(data=data, x=feature, ax=ax[1])
    ax[1].axvline(mean, color='r', linestyle='--', label='Mean')
    ax[1].axvline(median, color='g', linestyle='-', label='Median')
    ax[1].axvline(mode, color='b', linestyle='-.', label='Mode')
    ax[1].legend()

    plt.show()
```

```python
log_transformed_df = df[numerical_cols].apply(lambda x: np.log(x)) # reduce precision to two and apply log transformation
for f in numerical_cols:
    plot_tendency(data=log_transformed_df, feature=f, title=f"Tendency plot of {f}")
```

### 3.3 Dispersion
Measurement of how spread the data

```python
# measure central tendency
stds, variances, ranges, coefficient_variation = [], [], [], []
for f in numerical_cols:
    stds.append(df[f].std())
    variances.append(df[f].var())
    ranges.append(df[f].max() - df[f].min())
    coefficient_variation.append(df[f].std()/df[f].mean())


pd.DataFrame({
    'feature': numerical_cols,
    'std': stds,
    'variance': variances,
    'range': ranges,
    'coefficient variation': coefficient_variation,
})
```

The bigger the standard deviation the more varies the data for example in there are some movies that have outliers Trailer_views, or there are high uncertainty inside the data

## 4. Bivariate Analysis
Plot relationship between the data based on the Correlation Map.

### 4.1. Features vs `Collection` column

```python
def plot_correlation(data: pd.DataFrame, feature1: str, feature2: str, title: str):
    sns.lmplot(data=data, x=feature1, y=feature2)
    plt.title(title)
    plt.show()

plot_correlation(data=df, feature1='Budget', feature2='Collection', title="Budget vs Collection")
plot_correlation(data=df, feature1='Trailer_views', feature2='Collection', title="Trailer_Views vs Collection")
plot_correlation(data=df, feature1='Multiplex coverage', feature2='Collection', title="Multiplex coverage vs Collection")
plot_correlation(data=df, feature1='Critic_rating', feature2='Collection', title="Critic_rating vs Collection")
```

### 4.2 `Movie_length` or `Lead_Actor_Rating` or `Lead_Actress_rating` or `Director_rating` or `Producer_rating` vs `Production expense`

```python
plot_correlation(data=df, feature1='Movie_length', feature2='Production expense', title="Movie_length vs Production expense")
plot_correlation(data=df, feature1='Lead_ Actor_Rating', feature2='Production expense', title="Lead_Actor_Rating vs Production expense")
plot_correlation(data=df, feature1='Lead_Actress_rating', feature2='Production expense', title="Lead_Actress_Rating vs Production expense")
plot_correlation(data=df, feature1='Director_rating', feature2='Production expense', title="Director_rating vs Production expense")
plot_correlation(data=df, feature1='Producer_rating', feature2='Production expense', title="Producer_rating vs Production expense")
```

### 4.3  `Lead_Actor_Rating` or `Lead_Actress_rating` or `Director_rating` with `Producer_rating`

```python
plot_correlation(data=df, feature1='Lead_ Actor_Rating', feature2='Producer_rating', title="Lead_Actor_Rating vs Producer_rating")
plot_correlation(data=df, feature1='Lead_Actress_rating', feature2='Producer_rating', title="Lead_Actress_Rating vs Producer_rating")
plot_correlation(data=df, feature1='Director_rating', feature2='Producer_rating', title="Director_rating vs Producer_rating")
```

This phenomenon called multicollinearity as the correlation between independent variables makes the model difficult to estimate between relationship independent variables (feature) and dependent variable (label or target). We need to deal with this later in feature engineering.
